```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px #vamos criar gráfico com plotly nao com o matp
```

```python
#importação

#Tres linhas na esq > pasta > carregar
#variavel = pd.read_ ....

df_principal = pd.read_excel("/content/planilha_zerada.xlsx", sheet_name="Principal")

#dica: absoluta + botao direito na planilha no colab > copiar caminho
#ou escrever o nome da planilha
#Sheet vai carregar  acoluna

#head = mostra as linhas = datafra.head(linhas)
df_principal.head(10)
```

```python
#CARREGANDO A COLUNA AÇÕES

#obs? o nome sheet é o nome que está no excel

df_total_acoes = pd.read_excel("/content/planilha_zerada.xlsx", sheet_name="Total_de_acoes")
df_total_acoes.head(10)
```

```python
df_ticker = pd.read_excel("/content/planilha_zerada.xlsx", sheet_name="Ticker")
df_ticker.head(10)
```

```python
df_dadosgpt = pd.read_excel("/content/planilha_zerada.xlsx", sheet_name="chat_gpt")
df_dadosgpt.head(11)
```

```python
#descartando colunas desnecessárias e seleiconando só o qeu eu quero

df_principal = df_principal[["Ativo", "Data", "Último (R$)", "Var. Dia (%)"]].copy() #vai copiar ele mesmo >> pega o df que já existe, copiando e só com essas colunas ai e substituindo na variavel df_principal

#pq eu escrevi dataframe = dataframe[[]] < pq eu quero alterar ESTE MESMO DATAFRAME e pegar as colunas de lá por isso lista dentro de lista

df_principal
```

```python
#em python, caracteres especiais fazem diferença. Veja que algumas tem r$. vamos renomear as duas ultimas colunas.

#vamos sempre alterando o dataframe com rename ()
#df = df.rename( o que vou alterar ex: coluna, data, index{"nome da coluna que quero alterar": "novo nome, "nome da coluna que quero alterar 2": "novo nome"})
df_principal = df_principal.rename (columns = {"Último (R$)":"valor_final", "Var. Dia (%)": "var_dia_pct"}).copy()
df_principal


#sempre dar um .copy() > ele apaga o original e faz um novo
```

```python
#calculando a var_pct = df[novo nome para coluna] = df["coluna que quero pegar do csv"] + operação

#dividindo a variação por 100 e afins igual excel na tabella tratada

df_principal["var_pct"] = df_principal["var_dia_pct"] / 100 #criei agora
df_principal["valor_inicial"] = df_principal["valor_final"]/(df_principal["var_pct"] + 1) #mesmo calculo da do excel = valor% (l2 +1)


df_principal
```

```python
#usando um "procv"/vlookup no python = merge

#.merge()
#o uso é tipo inner join pense sempre em "tabela esq" e "direita". Pegar a aba PRINCIPAL e conectar com a aba TOTAL_ACOES

#meu dataframe = meudataframe.merge(dataframe que eu to buscando a info(dica importar isoladamente), o que tem de incomum nas tabelas com left_on= ou right_on=, indicar quem é a principal com how=left (indiquei que a esq é a principal) )

df_principal = df_principal.merge(df_total_acoes,left_on="Ativo", right_on="Código", how="left") #obs: total de ações é uma aba do excel| estou dizendo no how que DF_principal é a tabela principal e vou acrescentar a ela o pós () "df_total_acoes"
df_principal #obs se não tiver info - retorna null
```

```python
df_principal= df_principal.drop(columns=["Código"])
df_principal
```

```python
#criando a coluna variação em $

#calculo do excel: =(c2-m2)*m2

df_principal["variacao_rs"] = (df_principal["valor_final"] - df_principal["valor_inicial"]) * df_principal["Qtde. Teórica"]
df_principal
```

```python
#comando mais decorado para alterar a forma de mostrar

#para formatar todos os valores do tipo float - decoreba

pd.options.display.float_format = "{:.2f}".format
df_principal
```

```python
#convertendo qut teorica de float PARA > INT <

df_principal["Qtde. Teórica"] = df_principal["Qtde. Teórica"].astype(int)
df_principal
```

```python
#tirando o acento da qtd teorica
df_principal = df_principal.rename(columns={"Qtde. Teórica": "Qtd_teorica"}).copy()
df_principal
```

```python
#criando a coluna resultado subiu e desceu

#perguntei no gpt - transformando status em resultado e trazendo a formula do excel

#np.where(df['Valor'] > 0, 'SUBIU', np.where(df['Valor'] < 0, 'DESCEU', 'ESTÁVEL'))

df_principal['Resultado'] = np.where(df_principal['variacao_rs'] > 0, 'SUBIU', np.where(df_principal['variacao_rs'] < 0, 'DESCEU', 'ESTÁVEL'))

df_principal
```

```python
#parei em 42:24 - https://cursos.alura.com.br/imersoes/aulas/aula-03-manipulacao-de-dados-e-criacao-de-graficos-com-bibliotecas-python-c125?utm_source=devzapp&utm_medium=whatsapp&utm_content=aula-03&utm_campaign=imersaopython

df_principal = df_principal.merge(df_ticker,left_on="Ativo", right_on="Ticker", how="left") #obs: total de ações é uma aba do excel| estou dizendo no how que DF_principal é a tabela principal e vou acrescentar a ela o pós () "df_total_acoes"
df_principal= df_principal.drop(columns=["Ticker"])
df_principal
```

```python
#trazendo nome das ações
df_principal = df_principal.merge(df_dadosgpt,left_on="Nome", right_on="Empresa", how="left") #obs: total de ações é uma aba do excel| estou dizendo no how que DF_principal é a tabela principal e vou acrescentar a ela o pós () "df_total_acoes"
df_principal= df_principal.drop(columns=["Empresa"])

df_principal
```

```python
#alterando a coluna "nome" para EMPRESA

df_principal = df_principal.rename(columns={"nome": "Empresa"}).copy()
df_principal
```

```python
#alterando o nome e tirando ()
df_principal = df_principal.rename(columns={"Idade (Anos)": "Idade_Empresa"}).copy()
df_principal
```

```python
#criando a coluna cat_idade com CHATGPT


df_principal['cat_idade'] = np.select([df_principal['Idade_Empresa'] > 100, df_principal['Idade_Empresa'] < 50],
                                      ['Maior que 100 anos', 'Menos que 50'],
                                      default='Entre 50 e 100')
df_principal
```

```python
#fazendo as análises 47:40
```

```python
#usando gpt e fazendo a aba analises

# Calculando os valores da tabela
maior = df_principal['variacao_rs'].max() #max
var_dia = df_principal['variacao_rs'].diff().iloc[-1]
menor = df_principal['variacao_rs'].min() #min

#médias
varia_dia = df_principal['variacao_rs'].diff().mean() #.mean = media
media_geral = df_principal['variacao_rs'].mean()
media_subiram = df_principal[df_principal['variacao_rs'] > 0]['variacao_rs'].mean()
media_cairam = df_principal[df_principal['variacao_rs'] < 0]['variacao_rs'].mean()

# Imprimindo os resultados formatados
print(f"Maior\t R$ {maior:,.2f}")
print(f"Variação dia\t R$ {var_dia:,.2f}")
print(f"Menor\t R$ {menor:,.2f}")
print(f"Varia. Dia\t R$ {varia_dia:,.2f}")
print(f"Média geral\t R$ {media_geral:,.2f}")
print(f"Média dos que subiram\t R$ {media_subiram:,.2f}")
print(f"Média dos que cairam\t R$ {media_cairam:,.2f}")
```

```python
#analises de segmentos

#criando um df_principal só com os que subiram

df_principal_subiu = df_principal[df_principal['variacao_rs'] > 0].copy().assign(media_subiram=df_principal[df_principal['variacao_rs'] > 0]['variacao_rs'].mean())
df_principal_subiu
```

```python
#somando tudo com segmento - GROUP BY
#df novo = df que quero agrupar. GROUPBY (ONDE QUERO AGRUPAR)

#SE QUISER FAZER OPERACOES MAT: coloco [].sum por ex

#se eu der um groupby puro ele cria um verotr. Se colocarmos o RESET_INDEX => nós continuamos mexendo com o mesmo dataframe
df_analise_segmento = df_principal_subiu.groupby("Segmento")["variacao_rs"].sum().reset_index()
df_analise_segmento
```

```python
#analise de saldo
df_analise_saldo = df_principal.groupby("Resultado")["variacao_rs"].sum().reset_index()
df_analise_saldo
```
